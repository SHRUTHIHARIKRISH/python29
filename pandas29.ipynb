```python
import pandas as pd

# -----------------------------
# Step 1: Create the dataset
# -----------------------------
data = {
    "Student_ID": [101, 102, 103, 104, 105, 102],
    "Name": ["John", "Anna", "Peter", "Linda", "David", "Anna"],
    "Department": ["CS", "IT", "CS", "IT", "CS", "IT"],
    "Maths": [80, 90, None, 70, 60, 90],
    "Science": [85, 95, 75, None, 65, 95],
    "English": [78, 88, 70, 60, None, 88],
    "City": ["Chennai", "Chennai", "Madurai", "Chennai", "Madurai", "Chennai"]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 2: Basic understanding
# -----------------------------
print("First 5 rows:")
print(df.head())

print("\nDataset info:")
print(df.info())

print("\nShape of dataset:")
print(df.shape)

print("\nNull values in each column:")
print(df.isnull().sum())
print("\n" + "="*50)

# -----------------------------
# Step 3: Fill missing values
# -----------------------------
df["Maths"] = df["Maths"].fillna(0)
df["Science"] = df["Science"].fillna(0)
df["English"] = df["English"].fillna(0)

print("After filling missing values with 0:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 4: Find and remove duplicates
# -----------------------------
print("Duplicate Student_ID rows:")
print(df[df.duplicated(subset="Student_ID", keep=False)])

df = df.drop_duplicates(subset="Student_ID", keep="first")

print("\nAfter removing duplicate Student_ID:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 5: Create Total and Average
# -----------------------------
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

print("After adding Total and Average:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 6: Create Result using lambda
# -----------------------------
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")

print("After adding Result column:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 7: Conditional indexing
# -----------------------------
print("Students with Average > 75:")
print(df[df["Average"] > 75])

print("\nStudents who failed:")
print(df[df["Result"] == "Fail"])
print("\n" + "="*50)

# -----------------------------
# Step 8: loc and iloc examples
# -----------------------------
print("2nd student details using loc:")
print(df.loc[1])

print("\nName and Total columns:")
print(df.loc[:, ["Name", "Total"]])

print("\nFirst row using iloc:")
print(df.iloc[0])

print("\nFirst 3 rows and first 2 columns using iloc:")
print(df.iloc[0:3, 0:2])
print("\n" + "="*50)

# -----------------------------
# Step 9: Drop City column
# -----------------------------
df = df.drop("City", axis=1)

print("After dropping City column:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 10: Rename columns
# -----------------------------
df = df.rename(columns={
    "Maths": "Math_Score",
    "Science": "Science_Score"
})

print("After renaming columns:")
print(df)
print("\n" + "="*50)

# -----------------------------
# Step 11: Sort by Total descending
# -----------------------------
df_sorted = df.sort_values(by="Total", ascending=False)

print("Sorted by Total marks (highest to lowest):")
print(df_sorted)

print("\nTop 3 students:")
print(df_sorted.head(3))
```

```python
import pandas as pd

# -----------------------------
# Step 1: Create Dataset
# -----------------------------
data = {
    "Emp_ID": [201, 202, 203, 204, 205, 202],
    "Name": ["John", "Anna", "Peter", "Linda", "David", "Anna"],
    "Department": ["IT", "HR", "IT", "Finance", "IT", "HR"],
    "Salary": [60000, 45000, 70000, 50000, 55000, 45000],
    "Bonus": [5000, None, 7000, None, 3000, None],
    "Experience": [6, 3, 8, 5, 7, 3],
    "City": ["Chennai", "Chennai", "Madurai", "Chennai", "Madurai", "Chennai"]
}

df = pd.DataFrame(data)

print("Original Data:\n", df)
print("\n" + "="*50)

# -----------------------------
# Step 2: Check null values
# -----------------------------
print("Null values:\n", df.isnull().sum())
print("\n" + "="*50)

# -----------------------------
# Step 3: Fill missing Bonus with 0
# -----------------------------
df["Bonus"] = df["Bonus"].fillna(0)

# -----------------------------
# Step 4: Remove duplicates
# -----------------------------
df = df.drop_duplicates(subset="Emp_ID", keep="first")

# -----------------------------
# Step 5: Rename columns
# -----------------------------
df = df.rename(columns={
    "Emp_ID": "Employee_ID",
    "Salary": "Base_Salary"
})

# -----------------------------
# Step 6: Create Total Salary
# -----------------------------
df["Total_Salary"] = df["Base_Salary"] + df["Bonus"]

# -----------------------------
# Step 7: Lambda function (Senior/Junior)
# -----------------------------
df["Level"] = df["Experience"].apply(
    lambda x: "Senior" if x > 5 else "Junior"
)

# -----------------------------
# Step 8: Conditional filtering
# -----------------------------
high_salary = df[df["Base_Salary"] > 50000]
it_employees = df[df["Department"] == "IT"]

print("Employees with Salary > 50000:\n", high_salary)
print("\nIT Department Employees:\n", it_employees)
print("\n" + "="*50)

# -----------------------------
# Step 9: loc and iloc
# -----------------------------
print("Using loc (row 1):\n", df.loc[1])
print("\nUsing iloc (first row):\n", df.iloc[0])

print("\nSelect Name and Total Salary:\n", df.loc[:, ["Name", "Total_Salary"]])
print("\n" + "="*50)

# -----------------------------
# Step 10: Drop City column
# -----------------------------
df = df.drop("City", axis=1)

# -----------------------------
# Step 11: Highest paid employee
# -----------------------------
highest_paid = df[df["Total_Salary"] == df["Total_Salary"].max()]

print("Highest Paid Employee:\n", highest_paid)
print("\n" + "="*50)

# -----------------------------
# Final Data
# -----------------------------
print("Final Cleaned Data:\n", df)
```
