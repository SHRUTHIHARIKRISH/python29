Analyze marks of students using NumPy

```python
marks=np.array([45, 67, 89, 34, 90, 76, 88])
avg=np.average(marks)
high=np.max(marks)
low=np.min(marks)
add=marks+5
greatest=marks[marks>75]
new_marks=np.astype(marks, float)


print(avg)
print(high)
print(low)
print(add)
print(greatest)
print(new_marks)
```
