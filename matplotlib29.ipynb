```python
#Histogram
data = [10, 20, 20, 30, 30, 30, 40]
plt.hist(data, color=["YELLOW"])
plt.title("Histogram")

plt.show()
```

```python
import pandas as pd
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
df=pd.DataFrame(data)
plt.plot(df)
plt.show()
```

```python
from matplotlib.lines import lineStyles
import pandas as pd
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
df=pd.DataFrame(data)
plt.plot(df["Salary"], color="RED",linewidth="2",linestyle=":",marker="o",markerfacecolor="yellow")
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [10, 20, 30, 40, 100]
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```

```python
#pie chart
name= ["A", "B", "C"]
sizes = [40, 35, 25]
 #shows percentage on slices,"1.1f → show 1 number after decimal %% → print % symbol"
plt.pie(sizes,labels=name,autopct="%2.1f%%" ,explode=[0.1,0,0])
plt.title("PIE CHART")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = ["A","B","A","C","B","A"]
#countplot = count + plot

sns.countplot(x=data)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,30,40]
plt.scatter(x,y , color="RED")
plt.title="Scarrter plot"
plt.show()
```

```python
names = ["A","B","C"]
values = [10,20,15]
plt.bar(names,values)
plt.show()
```

MULTIVARIATE ANALYSIS
more than 2 variables

```python
x = [1,2,3]
y = [10,20,30]
size = [100, 300, 500]
plt.scatter(x,y,s=size)
#s is a parameter name (built into matplotlib)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import importlib

# Reload matplotlib.pyplot to ensure functions like xlabel are reset
importlib.reload(plt)

data={
    "Age": [22, 25, 30, 35, 28, 40],
    "Salary": [20000, 30000, 50000, 70000, 40000, 90000],
    "dept": ["HR", "IT", "Finance", "IT", "HR", "Finance"]
}
df = pd.DataFrame(data)

color_dept={"HR":"YELLOW","IT": "BLUE" , "Finance" : "ORANGE"}

for dept, color in color_dept.items():
    dept_color=df[df["dept"]== dept]
    # Corrected 's' parameter: use the 'Salary' column values for size, scaled down
    plt.scatter(dept_color["Age"],dept_color["Salary"],s =dept_color["Salary"]/100, color=color,label=dept)



plt.title("SCATTER PLOT")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
