```python
import matplotlib.pyplot as plt
```

```python
#Histogram
data = [10, 20, 20, 30, 30, 30, 40]
plt.hist(data, color=["YELLOW"])
plt.title("Histogram")

plt.show()
```

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y) #(1,4)(2,5)(3,6)
plt.grid()
plt.show()
```

```python
import pandas as pd
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
df=pd.DataFrame(data)
plt.plot(df)
plt.show()
```

```python
from matplotlib.lines import lineStyles
import pandas as pd
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
df=pd.DataFrame(data)
plt.plot(df["Salary"], color="RED",linewidth="2",linestyle=":",marker="o",markerfacecolor="yellow")
plt.show()
```
